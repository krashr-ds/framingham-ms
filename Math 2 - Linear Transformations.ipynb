## Graphing Linear Transformations using Quiver

First, we plot the basis vectors, i hat and j hat. 

These concepts are derived from the 3Blue1Brown video on Linear Transformations:
https://www.youtube.com/watch?v=kYB8IZa5AuE

i hat is a vector of length 1, that extends along the x axis (right).

j hat is a vector of length 1, that extends along the y axis (up).

I made them very small relative to the rest of the grid to give us space to scale and transform them.

```python
xp, yp = [0,0], [0,0]
xd, yd = [1,0], [0,1]

fig, ax = plt.subplots(figsize = (10,5))
ax.quiver(xp, yp, xd, yd, scale=30, headaxislength=2, headlength=2)
ax.axis([-30, 30, -30, 30])
plt.grid(True)
plt.show()
```

Scale i hat and j hat by the vector, [3,-2]

```python
xp, yp = np.array([0,0]), np.array([0,0])
yh, jh = np.array([1,0]), np.array([0,1])
transformation_matrix = np.array([[3, -1],
                                  [-2, 2]])
xd = yh.dot(transformation_matrix)
yd = jh.dot(transformation_matrix)

print("xd is now", xd)
print("yd is now", yd)

fig, ax = plt.subplots(figsize = (10,5))
ax.quiver(xp, yp, xd, yd, scale=50, color="blue", headaxislength=2, headlength=2)
ax.axis([-50, 50, -50, 50])
plt.grid(True)
plt.show()
```

Apply two transformations, then scale by a factor of 4.

```python
#begin at the origin
xp, yp = np.array([0,0]), np.array([0,0])
#initialize basis vectors
yh, jh = np.array([1,0]), np.array([0,1])

#transformations: 

m1 = np.array([[1, -2],
               [1, 0]])
m2 = np.array([[0, 2],
               [1, 0]])

m3 = m1.dot(m2)
xd = yh.dot(m3)
yd = jh.dot(m3)

print("m1 times m2 is", m3)
print("y hat is now", xd)
print("j hat is now", yd)

#scale by a factor of 4
xd = xd * 4
yd = yd * 4

print("xd has scaled to", xd)
print("yd has scaled to", yd)


fig, ax = plt.subplots(figsize = (10,5))
ax.quiver(xp, yp, xd, yd, scale=50, color="red", headaxislength=2, headlength=2)
ax.axis([-50, 50, -50, 50])
plt.grid(True)
plt.show()
```

A three dimensional example, with y hat (1,0,-1), j hat (1,1,0), and k hat (1,0,1)

Reference: https://youtu.be/rHLEWRxRGiM

```python
from mpl_toolkits.mplot3d import Axes3D
```

```python
fig = plt.figure(figsize=(10,5))
ax = fig.add_subplot(projection='3d')

# initialize the basis vectors
xp, yp, zp = np.array([0,0,0]), np.array([0,0,0]), np.array([0,0,0])
#initialize basis vectors
yh, jh, kh = np.array([1,0,-1]), np.array([1,1,0]), np.array([1,0,1])

#transform
transformation_matrix = np.array([[1, -2, 1],
                                  [2, -4, 4],
                                  [6, 7, -7]])

xd = yh.dot(transformation_matrix)
yd = jh.dot(transformation_matrix)
zd = kh.dot(transformation_matrix)

print("xd is ", xd)
print("yd is ", yd)
print("zd is ", zd)

# Can't change the size of the arrow heads this time, 
# because the 3D line plot implementation doesn't have these options.
ax.quiver(xp, yp, zp, xd, yd, zd, color="green")

ax.set_xlim([-9, 8])
ax.set_ylim([-9, 8])
ax.set_zlim([-9, 8])

plt.grid(True)
plt.show()
```
